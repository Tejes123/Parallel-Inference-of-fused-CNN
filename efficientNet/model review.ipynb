```python
import torch 
import torchvision 
```

```python
models = torchvision.models.list_models()
for model in models:
    print(model)
```

```python
model = torchvision.models.efficientnet_v2_s()
print(model)
```

```python
features_eff = model.features
for child in features_eff.named_children():
    print(child[0], ": ")
```

```python
model = torchvision.models.inception_v3()
print(model)
```

```python
for child in model.named_children():
    print(child[0], ": ", len(list(child[1].children())))
```
